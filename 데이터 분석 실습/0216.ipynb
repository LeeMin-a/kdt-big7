```python
index = 1
no = 1                  # 각 페이지 xpath 번호(1 ~ 5)
page = 1                # 현재 페이지
names = []
address = []
category = []

while 1:
    time.sleep(0.1)
    try:
        # time.sleep(0.1)
        names.append(driver.find_element(By.XPATH, f'//*[@id="info.search.place.list"]/li[{index}]/div[3]/strong/a[2]').text)
        address.append(driver.find_element(By.XPATH, f'//*[@id="info.search.place.list"]/li[{index}]/div[5]/div[2]/p[1]').text)
        category.append(driver.find_element(By.XPATH, f'//*[@id="info.search.place.list"]/li[{index}]/div[3]/span').text)
        # print(no, 'xpath | ', page, '페이지 | ', index, '현재 인덱스 | ', driver.find_element(By.XPATH, f'//*[@id="info.search.place.list"]/li[{index}]/div[3]/strong/a[2]').text)
        index += 1
    except:
        # 1페이지 더보기 버튼 있을때
        if driver.find_element(By.XPATH, '//*[@id="info.search.place.more"]').is_displayed():
            # time.sleep(0.1)
            # time.sleep(2)
            # print('case 1')
            driver.find_element(By.XPATH, '//*[@id="info.search.place.more"]').click()
            no += 1
            index = 1
            page += 1
            continue
        # 5배수 페이지(화살표 버튼 눌러 섹션 넘기기)
        elif no >= 5:
            # time.sleep(0.1)
            # time.sleep(2)
            # print('case 2')
            driver.find_element(By.XPATH, '//*[@id="info.search.page.next"]').click()
            no = 1
            page += 1
            index = 1
            continue
        # 다음 번호 페이지 넘기기
        elif driver.find_element(By.XPATH, '//*[@id="info.search.page.no' + str(no + 1) + '"]').is_displayed():
            # time.sleep(0.1)
            # time.sleep(2)
            # print('case 3')
            driver.find_element(By.XPATH, '//*[@id="info.search.page.no' + str(no + 1) + '"]').click()
            no += 1
            index = 1
            page += 1
            continue
        else:
            break
```

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
title = []
author = []
publisher = []
price = []
year = []
discount = []
count = 0
for i in range(1, 20):
    temp = 0
    temp_a = 0
    url_n = f'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=' + str(i) + '&cnt=1000&SortOrder=1'
    req = requests.get(url_n)
    content = req.content
    soup = BeautifulSoup(content, 'html.parser')
    for j in range(0, 50):
        count += 1
        title.append(soup.select('a.bo3')[j].text)
        print(title[count - 1], 'case1')
        print(j, j * 2, 'case2', sep=' | ')
        try:
            year.append(re.findall('\d\d\d\d년', soup.select('div.ss_book_list')[j * 2 + temp].select('ul')[0].text)[0])
        except:
            temp += 1
            print('temp : ', temp)
            year.append(re.findall('\d\d\d\d년', soup.select('div.ss_book_list')[j * 2 + temp].select('ul')[0].text)[0])
        print(year[count - 1], 'case3')
        discount.append(soup.select('span.ss_p2')[j].select('b')[0].text.replace(',', ''))
        print(discount[count - 1], 'case4')
        try:
            try:
                author.append(soup.select('div.ss_book_list')[j * 2 + temp_a].select('li')[2].select('a')[0].text)
                print(author[count - 1], 'case5')
                publisher.append(soup.select('div.ss_book_list')[j * 2 + temp_a].select('li')[2].select('a')[-1].text)
                print(publisher[count - 1], 'case6')
                price.append(int(soup.select('div.ss_book_list')[j * 2 + temp_a].select('ul')[0].select('li')[3].select('span')[0].text.replace(',', '')))
                print(price[count - 1], 'case7')
            except:
                author.append(soup.select('div.ss_book_list')[j * 2 + temp_a].select('li')[1].select('a')[0].text)
                print(author[count - 1], 'case8')
                publisher.append(soup.select('div.ss_book_list')[j * 2 + temp_a].select('li')[1].select('a')[-1].text)
                print(publisher[count - 1], 'case9')
                price.append(int(soup.select('div.ss_book_list')[j * 2 + temp_a].select('ul')[0].select('li')[2].select('span')[0].text.replace(',', '')))
                print(price[count - 1], 'case10')
        except:
            temp_a += 1
            try:
                author.append(soup.select('div.ss_book_list')[j * 2 + temp_a].select('li')[2].select('a')[0].text)
                print(author[count - 1], 'case11')
                publisher.append(soup.select('div.ss_book_list')[j * 2 + temp_a].select('li')[2].select('a')[-1].text)
                print(publisher[count - 1], 'case12')
                price.append(int(soup.select('div.ss_book_list')[j * 2 + temp_a].select('ul')[0].select('li')[3].select('span')[0].text.replace(',', '')))
                print(price[count - 1], 'case13')
            except:
                author.append(soup.select('div.ss_book_list')[j * 2 + temp_a].select('li')[1].select('a')[0].text)
                print(author[count - 1], 'case14')
                publisher.append(soup.select('div.ss_book_list')[j * 2 + temp_a].select('li')[1].select('a')[-1].text)
                print(publisher[count - 1], 'case15')
                price.append(int(soup.select('div.ss_book_list')[j * 2 + temp_a].select('ul')[0].select('li')[2].select('span')[0].text.replace(',', '')))
                print(price[count - 1], 'case16')
        print(count, '번째 책')
    print(i)
```

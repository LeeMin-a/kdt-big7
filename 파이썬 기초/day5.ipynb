* 데이터 이산화
    * 연속형 데이터->범주형 데이터
    * 기계학습에서 사용
        * 연속형 데이터
        * 범주형 데이터
            * 이산 데이터(binary data)
                * 둘 중 하나로 나뉘는 것
                * True, False
            * 고려해야 할 경우의 수가 적어짐

* 결측치 처리
    * 0-fill
    * mean, median
    * 삭제(선호x)
        * 너무 많이 삭제하면 신뢰도 보장 불가
* 이상치 처리
    * 제거/대체
* 표준화, 정규화
    * 표준화
        * 정규분포의 속성을 갖도록 데이터 분포 조절
        * 데이터 범위 제한 없음
        * z-score
    * 정규화
        * 데이터 분포 유지
        * 데이터 범위 제한(0~1)
        * min-max scaling

```python
import pandas as pd

media = pd.read_csv('./시청자미디어재단_체험형미디어교육_20240105.csv', encoding='cp949')
```

```python
media['센터명'] = media['센터명'].apply(lambda x: x[:2])
```

```python
# media.fillna(0)    # 0으로 결측치 채움
media.dropna(how='all', axis=0, inplace=True)    # row기준 모든값이 결측치일때 해당 row 삭제. any : 결측치가 한개라도 있으면 모두 삭제
```

```python
translate = {'유아' : '미취학',
'5~7세' : '미취학',
'일반시민' : '일반시민',
'초등생 이상' : '일반시민',
'중등' : '중등'}
```

```python
translate['5~7세']
```

```python
media['교육대상구분'] = [translate[m] for m in media['교육대상구분']]
```

```python
def chk_cat(x):
    if '더빙'in x:
        return '더빙'
    elif any([i in x for i in['뉴스', '라디오', '라이브커머스']]):
        return '컨텐츠'
    elif '크로마키' in x:
        return '크로마키'
    else:
        return x
```

```python
media['강좌내용'] = media['강좌내용'].apply(lambda x: chk_cat(x))
```

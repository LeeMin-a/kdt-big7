```python
# 업종별로 각 기술스택이 몇 번 등장하는지 카운트하는 딕셔너리 초기화
tech_count_by_category = {}

# 각 업종별로 기술스택 카운트 진행
for index, row in df_count.iterrows():
    category = row['업종분류']
    techs = row['이용하는 기술스택']
    
    if pd.isna(techs):  # NaN 값인 경우 건너뜀
        continue
    
    tech_list = techs.split(', ')
    
    if category not in tech_count_by_category:
        tech_count_by_category[category] = {}
    
    for tech in tech_list:
        tech_count_by_category[category][tech] = tech_count_by_category[category].get(tech, 0) + 1

# 결과 출력
for category, tech_counts in tech_count_by_category.items():
    print(f"업종분류: {category}")
    for tech, count in tech_counts.items():
        print(f" - {tech}: {count}번")
    print()
```

```python
counts = []
for techs in df_count['이용하는 기술스택']:
    if pd.isna(techs):  # NaN 값인 경우 건너뜀
        counts.append(0)
    else:
        count = sum(1 for tech in arr if tech in techs.split(', '))
        counts.append(count)

# 결과를 데이터프레임에 추가
df_count['기술스택 카운트'] = counts

df_count
```

```python
stack_counts = df_count.groupby('업종분류')[[]].sum()
stack_counts
```

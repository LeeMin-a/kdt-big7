```python
s_date = []
e_date = []
for d in soup.select('dl.item.term'):
    temp = re.findall('\d+-\d+-\d+', d.text)
    if temp ==[]:
        continue
    else:
        s_date.append(temp[0])
        e_date.append(temp[1])
```

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import time
company = []
job = []
year = []
category = []
edu = []
location = []
for i in range(0, 1261, 30):
    url = f'https://search.incruit.com/list/search.asp?col=job&kw=%b5%a5%c0%cc%c5%cd&startno={i}&psize=30'
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content, 'html.parser')
    for j in range(len(soup.select('div.cl_md'))):
        time.sleep(.1)
        company.append(soup.select('div.cl_top')[j * 3].select('a.cpname')[0].text)
        job.append(soup.select('div.cl_top')[j * 3 + 1].select('a')[0].text)
        year.append(soup.select('div.cl_md')[j].select('span')[0].text)
        category.append(soup.select('div.cl_md')[j].select('span')[3].text)
        edu.append(soup.select('div.cl_md')[6].select('span')[1].text)
        location.append(soup.select('div.cl_md')[6].select('span')[2].text)
```

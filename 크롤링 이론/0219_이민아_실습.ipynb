```python
import re
```

```python
region = []
for address in school_df['주소']:
    # 주소에서 마지막 단어(공백 기준) 제거
    address = ' '.join(address.split()[:-1])
    
    # 주소에서 '길'로 끝나는 단어 제거
    address = re.sub(r'\b\w+길\b', '', address)
    
    # 숫자로 끝나는 단어는 숫자를 제거
    address = re.sub(r'\d+$', '', address)
    
    # 주소에서 '동' 정보 추출
    dongs = re.findall(r'\b(\w+동\w*)\b', address)
    
    if dongs:
        region.extend(dongs)
    else:
        # '동'이 없는 경우 '구' 정보 추출
        gu = re.search(r'\b(\w+구)\b', address)
        if gu:
            region.append(gu.group(1))

# 결과 출력
for i in region:
    print(i)
len(region)
```
